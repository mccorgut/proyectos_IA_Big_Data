## Distribuciones normales

En este bloque nos vamos a familiarizar con las distribuciones normales, dibujando las dos que vamos a utilizar en el ejercicio que son estas:

- Distribución normal con media 1.000 y desviación estándar 1
- Distribución normal con media 1.000 y desviación estándar 4

La primera se da hecha, y se pide modificar el código para generar la segunda.

### Representación de la 1º distribución normal

```python
# Define los parámetros de la distribución
muA = 1000
sigmaA = 1

# Genera un rango de valores x para la distribución
x = np.linspace(muA - 3*sigmaA, muA + 3*sigmaA, 100)

# Calcula la función de densidad de probabilidad (PDF) para cada valor x
pdf = stats.norm.pdf(x, muA, sigmaA)

# Dibuja la distribución
plt.plot(x, pdf, label='Distribución Normal (muA = 1000, sigmaA = 1)')
plt.title('Distribución Normal')
plt.xlabel('Valores')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.show()
```

### Representación de la 2ª distribución normal

```python
# TODO Hacer aquí la representación de la segunda distribución normal.
muB = 1000
sigmaB = 4

x = np.linspace(muB - 3*sigmaB, muB + 3*sigmaB, 100)

pdf = stats.norm.pdf(x, muB, sigmaB)

plt.plot(x, pdf, label='Distribución Normal (muB = 1000, sigmaB = 4)')
plt.title('Distribución Normal')
plt.xlabel('Valores')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.show()
```

## Generación de datos

A continuación se crean dos conjuntos de datos aleatorios que representan la distribución de embotellado de dos embotelladoras (datosA y datosB), luego, los combina en una sola matriz (datos) para su posterior análisis.

Para generar el conjunto de datos, datosA, se utiliza una función de distribución normal tomando datos aleatorios, y se hace con np.random.normal muA y sigmaA representan la media y la desviación estándar de la distribución normal. En resumen se generan 1000 valores aleatorios provenientes de una distribución normal con una media de 1000 y una desviación estándar de 1, que se almacenan en la variable datosA y de forma similar para datosB



### Generación de datos embotelladora 1

```python
muA, sigmaA = 1000, 1 # media y desviación estándar embotelladora A
datosA = np.random.normal(muA, sigmaA, 1000)
```

#### Representación de datos de la embotelladora 1

```python
# Vamos a dibujar un histograma con los datos obtenidos y comprobamos su
# apariencia sobre los de la primera normal dibujada anteriormente para
# comprobar que siguen el patrón.
plt.hist(datosA, bins=30, density=True, label='Datos Aleatorios')
plt.title('Distribución Normal (muA = 1000, sigmaA = 1)')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

### Generación de datos embotelladora 2

```python
muB, sigmaB = 1000, 4 # media y desviación estándar embotelladora B

# TODO Completar el código para la segunda embotelladora.

datosB = np.random.normal(muB, sigmaB, 1000)
```

#### Representación de datos de la embotelladora 2

```python
# TODO Representa los datos de la embotelladora 2 de forma similar a como se
# hizo para la 1.
plt.hist(datosB, bins=30, density=True, label='Datos Aleatorios')
plt.title('Distribución Normal (muB = 1000, sigmaB = 4)')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

### Combinar datos embotelladora 1 y 2

```python
datos = np.column_stack((datosA, datosB)) # ¿Qué hace esto?
datos
```
