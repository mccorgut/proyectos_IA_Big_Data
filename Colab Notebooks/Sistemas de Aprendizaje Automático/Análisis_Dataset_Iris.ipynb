## 1. Carga del dataset

## 4. Actividades a realizar sobre el datset
Completa las siguientes actividades utilizando el dataset:

1. **Crear un gráfico que muestre la distribución de la longitud del pétalo por especie.**
   - Pista: Usa `sns.histplot` con `hue` para diferenciar las especies.

2. **Calcular el promedio del ancho del sépalo por especie.**
   - Pista: Usa `groupby` y `mean`.

3. **¿Qué especie tiene las flores con la mayor longitud promedio del pétalo?**

4. **Crear una visualización que combine al menos tres variables.**

## 5. Soluciones
A continuación, escribe tus soluciones a las tareas propuestas.

```python
# Escribe tu solución aquí para la actividad 1

# Distribución de la longitud del pétalo por especie
sns.histplot(df, x='petal_length', hue='species', kde=True, palette='Set2')
plt.title('Distribución de la Longitud del Pétalo por Especie')
plt.show()
```

```python
# Escribe tu solución aquí para la actividad 2

# Promedio del ancho del sépalo por especie
promedio_ancho_sepalo = df.groupby('species')['sepal_width'].mean()
print(promedio_ancho_sepalo)
```

```python
# Escribe tu solución aquí para la actividad 3

# Especie con la mayor longitud promedio del pétalo
longitud_promedio_petal = df.groupby('species')['petal_length'].mean()
especie_max_petal = longitud_promedio_petal.idxmax()
print(f'La especie con la mayor longitud promedio del pétalo es: {especie_max_petal}')
```

```python
# Escribe tu solución aquí para la actividad 4

# Visualización combinada de tres variables
sns.scatterplot(data=df, x='sepal_length', y='petal_width', hue='species', palette='Set2')
plt.title('Relación entre Longitud del Sépalo, Ancho del Pétalo y Especie')
plt.show()
```
